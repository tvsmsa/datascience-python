### Делим признаки на числовые и категориальные

```python
cat_columns = [] 
num_columns = [] 

for column_name in df.columns: 
    if (df[column_name].dtypes == object):
        cat_columns +=[column_name] 
    else:
        num_columns +=[column_name] 

print(' Категориальные данные:\t ',cat_columns, '\n Число столблцов = ',len(cat_columns))

print(' Числовые данные:\t ',  num_columns, '\n Число столблцов = ',len(num_columns))
```

# Анализ числовых признаков

```python
df[['index', 'Publishing Year', 'Book_average_rating', 'Book_ratings_count', 'gross sales', 'publisher revenue', 'sale price', 'sales rank', 'units sold']]  # Посмотрим на числовые признаки
```

```python
# Выведем стандартные статистические параметры
df.describe()
```

```python
#Столбец index дублирует информацию из других столбцов, следовательно его можно удалить. 
df = df.drop('index', axis=1)
#То же самое и для sales rank, этот столбец дублирует units sold
df = df.drop('sales rank',axis=1)
df 
```

```python
num_columns.remove('index') #Удалим столбцы из num_colums
num_columns.remove('sales rank')
```

```python
# Посмотрим корреляции числовых столбцов
df[num_columns].corr().style.background_gradient(cmap = 'YlOrRd', # наш цветовой стиль
                                    vmin = -1, # указываем нижний диапазон
                                    vmax=1) # указываем верхний диапазон
```

Наибольшую корреляцию имеет gross sales (продажи) с publisher revenue (прибыль издателя). Также видна зависимости между продажами и рейтингом книги (Book_ratings_count). Чем выше рейтинг книги, тем лучше продажи. Год издания не имеет значительной связи с другими признаками.

# Визуализация данных 

Рассмотрим гистограммы (распределение) значений по каждому числовому признаку 

```python
width = 2
height = int(np.ceil(len(num_columns) / width))
fig, ax = plt.subplots(nrows=height, ncols=width, figsize=(15, 8))

for idx, column_name in enumerate(num_columns):
    plt.subplot(height, width, idx + 1)

    sns.histplot(data=df, x=column_name, bins=20)
# Признаки в расположены данном порядке: 'Publishing Year', 'Book_average_rating', 'Book_ratings_count', 
# 'gross sales', 'publisher revenue', 'sale price', 'units sold'
```

```python
# Рассмотрим колонку 'gross sales' в логарифмическом масштабе.
plt.figure(figsize=(10,6))
sns.histplot(data=df, 
             x='gross sales',
             bins = 20, 
             log_scale = True);  # Логарифмический масштаб (для очень больших диапазонов)
```

```python
# Рассмотрим признак units sold подробнее

plt.figure(figsize=(8,4))
sns.histplot(data=df,
        x='units sold',
        bins = 10,
        log_scale = True);
```

## Построим скатерограмму, которая показывает зависимость рейтинга автора и общих продаж от жанра

```python
plt.style.use('bmh')
fig, ax = plt.subplots(1, figsize=(10, 6))
sns.scatterplot(data=df,
                x='gross sales', y='Author_Rating',
                hue='genre', 
                )
```

## Построим скатерограмму, которая показывает зависимость общих продаж и издателя от жанра

```python
plt.style.use('bmh')
fig, ax = plt.subplots(1, figsize=(15, 8))
sns.scatterplot(data=df,
                x='units sold', y='Publisher',
                hue='genre', 
                ax=ax)
```

```python
plt.style.use('bmh')
fig, ax = plt.subplots(1, figsize=(15, 8))
sns.scatterplot(data=df,
                x='publisher revenue', y='Publisher',
                hue='genre', 
                ax=ax)
```

Больше всего издаёт книг Amazon и в целом продаёт больше других, однако другие издатели имеют книги, которые принесли больше прибыли.

 ## Рассмотрим графики распpеделения категориальных переменных в завмсимоти от 'Status'

####  Распределение по издателям

```python
fig, ax = plt.subplots(1, figsize=(15, 7))
sns.countplot(data=df, x='Publisher', hue='genre',order=df['Publisher'].value_counts().index)
ax.tick_params(axis='x', rotation=90)
```

Все издатели больше всего продают жанровую литературу. За исключением HarperCollins Christian Publishing.

#### Распределение по моделям процессора 

```python
fig, ax = plt.subplots(1, figsize=(15, 7))
sns.countplot(data=df, x='Author_Rating', hue='genre', order=df['Author_Rating'].value_counts().index)
ax.tick_params(axis='x', rotation=90)
```

Авторы внезависимости от рейтинга больше всего пишут жанровую литературу, с нон-фикшн на втором месте.
