```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
vyb_file = 'parlament 2.xlsx'
df = pd.read_excel(vyb_file)
```

```python
#df = df.drop(columns=["Год кампании", "Название кампании", 
 #                     "6.1. Количество кандидатов, участвующих в выборах", "6.2.Кол-во самовыдвиженцев",
  #                    "6.3.Кол-во незарегистрированных кандидатов",
   #                   "ЛДПР голоса", "КПРФ голоса", "СР голоса", "Новые люди",
    #                  "Зеленая альтернатива"
     #       ])
df = df.drop(columns=["Год кампании", "Название кампании"])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляция')
plt.show()
```
